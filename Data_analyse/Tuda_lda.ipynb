```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from os import listdir
from os.path import isfile, join
```

# Sequential

```python

path =  '/hpc/comco/Sarah.M/Tuda_project/results/sequential/lda/'

onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
gamma_lda = [element for element in onlyfiles if 'gamma' in element]
acc_star_lda = [element for element in onlyfiles if 'accuracy_star' in element]
```

```python
for file in acc_star_lda:
    acc = pd.read_csv(path + file)
    plt.figure()
    plt.plot(acc)
    plt.ylim(0.4, 0.8)
    plt.plot((0,120),(0.5,0.5),'r:')
    plt.title(f'subject {file[-6:-4]}')
```

```python
for i, file in enumerate(gamma_lda):
    acc = pd.read_csv(path + file)
    plt.figure(figsize=(20, 3))
    for trial in range(10):
        plt.subplot(1,10,trial+1)
        plt.plot(acc[trial*120:(trial+1)*120])
        if i== 0:
            plt.title(f'trial #{trial}')
        if trial ==0:
            plt.ylabel(f'subject {file[-6:-4]}')
    #plt.ylim(0.4, 0.8)
    #plt.plot((0,120),(0.5,0.5),'r:')
```
